```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import RMSprop, Adam

# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/MyungKyuYi/AI-class/refs/heads/main/diabetes.csv")

df
```

```python
# ---------------- 회귀: BMI 예측 ----------------

# Feature/Target 분리
X_reg = df.drop(['BMI'], axis=1).values
y_reg = df['BMI'].values

# 데이터 분할
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# 스케일링
scaler_reg = StandardScaler()
X_reg_train = scaler_reg.fit_transform(X_reg_train)
X_reg_test = scaler_reg.transform(X_reg_test)
```

```python
# 회귀 모델 생성
reg_model = Sequential([
    Input(shape=(X_reg_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1)
])

reg_model.compile(optimizer=RMSprop(), loss='mse', metrics=['mse'])

# 학습
EPOCHS_REG = 20
history_reg = reg_model.fit(X_reg_train, y_reg_train, epochs=EPOCHS_REG, batch_size=32, validation_data=(X_reg_test, y_reg_test))

# 회귀 학습 결과 시각화
plt.figure(figsize=(8,6))
plt.plot(history_reg.history['loss'], label='Training Loss')
plt.plot(history_reg.history['val_loss'], label='Validation Loss', linestyle='--')
plt.title('BMI Regression Loss (TensorFlow)')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ---------------- 분류: Outcome 예측 ----------------

# Feature/Target 분리
X_cls = df.drop(['Outcome'], axis=1).values
y_cls = df['Outcome'].values

# 데이터 분할
X_cls_train, X_cls_test, y_cls_train, y_cls_test = train_test_split(X_cls, y_cls, test_size=0.2, random_state=42)

# 스케일링
scaler_cls = StandardScaler()
X_cls_train = scaler_cls.fit_transform(X_cls_train)
X_cls_test = scaler_cls.transform(X_cls_test)
```

```python
# 원-핫 인코딩
y_cls_train_cat = tf.keras.utils.to_categorical(y_cls_train, num_classes=2)
y_cls_test_cat = tf.keras.utils.to_categorical(y_cls_test, num_classes=2)

# 분류 모델 생성
cls_model = Sequential([
    Input(shape=(X_cls_train.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(2, activation='softmax')
])

cls_model.compile(optimizer=Adam(learning_rate=0.003), loss='categorical_crossentropy', metrics=['accuracy'])

# 학습
EPOCHS_CLS = 20
history_cls = cls_model.fit(X_cls_train, y_cls_train_cat, epochs=EPOCHS_CLS, batch_size=32, validation_data=(X_cls_test, y_cls_test_cat))

# 분류 학습 결과 시각화
fig, ax = plt.subplots(1, 2, figsize=(14,5))

# Loss
ax[0].plot(history_cls.history['loss'], label='Training Loss')
ax[0].plot(history_cls.history['val_loss'], label='Validation Loss', linestyle='--')
ax[0].set_title('Outcome Classification Loss (TensorFlow)')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].legend()
ax[0].grid(True)
```

```python
# Accuracy
ax[1].plot(history_cls.history['accuracy'], label='Training Accuracy')
ax[1].plot(history_cls.history['val_accuracy'], label='Validation Accuracy', linestyle='--')
ax[1].set_title('Outcome Classification Accuracy (TensorFlow)')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy')
ax[1].legend()
ax[1].grid(True)

plt.show()

# 평가
y_cls_pred = cls_model.predict(X_cls_test)
y_cls_pred_labels = np.argmax(y_cls_pred, axis=1)

print(classification_report(y_cls_test, y_cls_pred_labels))
print(confusion_matrix(y_cls_test, y_cls_pred_labels))
```
