```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam

# ------------------- 데이터 로딩 -------------------

df = pd.read_csv("https://raw.githubusercontent.com/MyungKyuYi/AI-class/refs/heads/main/abalone.csv")
df
     
```

```python
# ------------------- 데이터 전처리 -------------------

# 성별(Sex) 컬럼 Label Encoding
encoder = LabelEncoder()
df['Sex'] = encoder.fit_transform(df['Sex'])

print("\nEncoded DataFrame:\n", df.head())
```

```python
# ------------------- Feature / Target 분리 -------------------

X = df.drop('Rings', axis=1).values  # 입력 변수 (길이, 무게 등)
y = df['Rings'].values               # 타겟 변수 (나이 예측)

print("\nFeature shape:", X.shape)
print("Target shape:", y.shape)
```

```python

# ------------------- 데이터 분할 -------------------

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("\nTrain/Test size:")
print("X_train:", X_train.shape, "X_test:", X_test.shape)
print("y_train:", y_train.shape, "y_test:", y_test.shape)

# ------------------- 데이터 스케일링 -------------------

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```
